```julia
using Plots
x = 1:10; y = rand(10); # These are the plotting data
plot(x,y)
```

```julia
using PyPlot
x = range(0,stop=2*pi,length=1000); y = sin.(3*x + 4*cos.(2*x))
plot(x, y, color="red", linewidth=2.0, linestyle="--")
```
